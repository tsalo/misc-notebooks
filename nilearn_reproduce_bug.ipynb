```python
%matplotlib inline
from nilearn.glm.first_level import make_first_level_design_matrix
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

data = [[1, 1, 'cond1']] # duplicate event at 0 secs not counted
events = pd.DataFrame(data, columns=['onset','duration','trial_type'])
X0 = make_first_level_design_matrix(np.arange(0, 10, 0.1), events=events, hrf_model=None)

data = [[1, 1, 'cond1'], [1, 1, 'cond1']] # duplicate event ~ 0 secs is now counted
events = pd.DataFrame(data, columns=['onset','duration','trial_type'])
X1 = make_first_level_design_matrix(np.arange(0, 10, 0.1), events=events, hrf_model=None)

fig, ax = plt.subplots(figsize=(16, 8))
ax.plot(X0["cond1"], label="Ignore", alpha=0.5)
ax.plot(X1["cond1"], label="Treat as Amplitude Modulator", alpha=0.5)
ax.legend()
plt.savefig('show_make_first_level_design_matrix_oddity.jpg')
```

```python
import pandas as pd

df = pd.read_table("/Users/taylor/Downloads/temp.tsv")
```

```python
def get_run(string):
    parts = string.split("_")
    run_str = [p for p in parts if p.startswith("run")][0]
    return run_str

def get_echo(string):
    parts = string.split("_")
    echo_str = [p for p in parts if p.startswith("echo")][0]
    return echo_str

def get_site(string):
    if "CN" in string:
        site = "CN"
    elif "FR" in string:
        site = "FR"
    elif "EN" in string:
        site = "EN"
    return site

df["run"] = df["file"].apply(get_run)
df["echo"] = df["file"].apply(get_echo)
df["site"] = df["subject_id"].apply(get_site)
df
```

```python
temp = df.groupby(["subject_id", "run"])
temp.mean()
```

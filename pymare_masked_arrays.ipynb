```python
import numpy as np
from numpy import ma
from pymare import core, estimators
```

```python
n_studies = 110
v = np.random.normal(5, 2, size=(n_studies))
y = np.random.normal(10, 3, size=(n_studies))
X = np.ones(n_studies)
n = np.random.randint(5, 50, size=n_studies)

mask = np.zeros(n_studies, int)
mask[:10] = 1
mask = mask.astype(bool)
mv = ma.masked_array(v, mask)
mX = ma.masked_array(X, mask)
my = ma.masked_array(y, mask)
mn = ma.masked_array(n, mask)
```

```python
dataset_full = core.Dataset(v=v, X=X, y=y, n=n, add_intercept=False)
dataset_red = core.Dataset(v=v[~mask], X=X[~mask], y=y[~mask], n=n[~mask], add_intercept=False)
dataset_masked = core.Dataset(v=mv, X=mX, y=my, n=mn, add_intercept=False)
dataset_masked.X = mX[:, None]
dataset_masked.v = mv[:, None]
dataset_masked.y = my[:, None]
dataset_masked.n = mn[:, None]
```

```python
est = estimators.WeightedLeastSquares().fit(dataset_full)
results = est.summary()
print(results.to_df())
```

```python
est = estimators.WeightedLeastSquares().fit(dataset_red)
results = est.summary()
print(results.to_df())
```

```python
est = estimators.WeightedLeastSquares().fit(dataset_masked)
results = est.summary()
print(results.to_df())
```

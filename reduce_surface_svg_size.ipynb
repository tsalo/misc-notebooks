```python
%matplotlib inline
import matplotlib.pyplot as plt
from nilearn import datasets, plotting

fig, axes = plt.subplots(figsize=(4, 4), ncols=2, nrows=2, subplot_kw={"projection": "3d"})

fsaverage = datasets.fetch_surf_fsaverage()
plotting.plot_surf(
    fsaverage.infl_right, 
    hemi='right',
    title='Surface right hemisphere', 
    colorbar=True,
    threshold=1., 
    bg_map=fsaverage.sulc_right,
    axes=axes[0, 0],
    figure=fig,
)
plotting.plot_surf(
    fsaverage.infl_right, 
    hemi='right',
    title='Surface right hemisphere', 
    colorbar=True,
    threshold=1., 
    bg_map=fsaverage.sulc_right,
    axes=axes[0, 1],
    figure=fig,
)
plotting.plot_surf(
    fsaverage.infl_right, 
    hemi='right',
    title='Surface right hemisphere', 
    colorbar=True,
    threshold=1., 
    bg_map=fsaverage.sulc_right,
    axes=axes[1, 0],
    figure=fig,
)
plotting.plot_surf(
    fsaverage.infl_right, 
    hemi='right',
    title='Surface right hemisphere', 
    colorbar=True,
    threshold=1., 
    bg_map=fsaverage.sulc_right,
    axes=axes[1, 1],
    figure=fig,
)
fig.tight_layout()
fig.show()
```

```python
from svgutils.transform import fromstring
from niworkflows.viz.utils import compose_view, extract_svg

obj = type("", (), {})()
subobj = type("", (), {})()
subobj.figure = fig
obj.frame_axes = subobj
svg = extract_svg(obj, dpi=300, compress=False)
svg = fromstring(svg)
svg.save("temp2.svg")
```
